```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as font_manager
```

# Experiment 3

```python
times8x8 = np.asarray([1674, 2435, 3491, 3796, 4149, 4462, 5061, 5439, 5607, 6070])
times4x4 = np.asarray([3623, 6822, 8950, 10796, 13802, 16577, 20010, 22422, 23662, 26561])
times2x2 = np.asarray([13789, 28374, 42512, 62150, 77400, 99993, 132570, 160585, 179805, 179805])
times2x1 = np.asarray([26302, 54536, 97234, 157146, 242071, 345753, 462870, 489584, 577950, 653306])
times1x1 = np.asarray([60413, 141930, 216636, 373450, 652529, 941261, 1.3362e+06, 1.52572e+06, 1.81182e+06, 2.1149e+06])
x = np.linspace(0, 9, 10)
xx = np.linspace(0, 9, 10)
fig, ax = plt.subplots(figsize=(16, 12))
ax.plot(x, (times8x8/1000)/60, label='8x8', marker='p', linestyle='-', linewidth=5, markersize=15)  # Plot some data on the axes.
ax.plot(x, (times4x4/1000)/60, label='4x4', marker='o', linestyle='--', linewidth=5, markersize=15)  # Plot more data on the axes...
ax.plot(x, (times2x2/1000)/60, label='2x2', marker='d', linestyle='-', linewidth=5, markersize=15)  # ... and some more.
ax.plot(x, (times2x1/1000)/60, label='2x1', marker='v', linestyle='-', linewidth=5, markersize=15)  # ... and some more.
ax.plot(xx,(times1x1/1000)/60, label='1x1', marker='s', linestyle='-', linewidth=5, markersize=15)  # ... and some more.
ax.set_xlabel('Iteration Step', fontsize=40, fontweight="bold")  # Add an x-label to the axes.
plt.xticks(fontsize=40, fontweight="bold")
ax.set_ylabel('Step time in minutes', fontsize=40, fontweight="bold")  # Add a y-label to the axes.
plt.yticks(fontsize=40, fontweight="bold")
ax.set_title("Performance Comparision of 3200 MTs \n on Different Subdivisions", fontsize=40, fontweight="bold", pad=22)  # Add a title to the axes.
leg = ax.legend(prop={'size': 36, 'weight': 'bold'});  # Add a legend.
leg.set_title('Subdivisions',prop={'size':40, 'weight': 'bold'})
fig.savefig('10_step_stackup_nolog_minutes.eps', bbox_inches = "tight")
```

```python
fig, ax = plt.subplots(figsize=(16, 12))
ax.plot(x, np.log((times8x8/1000)/60), label='8x8', marker='p', linestyle='-', linewidth=5, markersize=15)  # Plot some data on the axes.
ax.plot(x, np.log((times4x4/1000)/60), label='4x4', marker='o', linestyle='--', linewidth=5, markersize=15)  # Plot more data on the axes...
ax.plot(x, np.log((times2x2/1000)/60), label='2x2', marker='d', linestyle='-', linewidth=5, markersize=15)  # ... and some more.
ax.plot(x, np.log((times2x1/1000)/60), label='2x1', marker='v', linestyle='-', linewidth=5, markersize=15)  # ... and some more.
ax.plot(xx,np.log((times1x1/1000)/60), label='1x1', marker='s', linestyle='-', linewidth=5, markersize=15)  # ... and some more.
ax.set_xlabel('Iteration Step', fontsize=40, fontweight="bold")  # Add an x-label to the axes.
plt.xticks(fontsize=40, fontweight="bold")
ax.set_ylabel('Step time in LN(minutes)', fontsize=40, fontweight="bold")  # Add a y-label to the axes.
plt.yticks(fontsize=40, fontweight="bold")
ax.set_title("LN Scaled Performance Comparision\n of 3200 MTs on Different Subdivisions", fontsize=40, fontweight="bold", pad=22)  # Add a title to the axes.
leg = ax.legend(loc='upper left', prop={'size': 23, 'weight': 'bold'});  # Add a legend.
leg.set_title('Subdivisions',prop={'size':23, 'weight': 'bold'})
fig.savefig('10_step_stackup_log_minutes.eps', bbox_inches = "tight")
```
