```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

#mean absolute scaled error    
def mase(actual : np.ndarray, predicted : np.ndarray):

    forecast_error = np.mean(np.abs(actual - predicted))
    naive_forecast = np.mean(np.abs(np.diff(actual)))
    mase = forecast_error / naive_forecast
    return mase

def fit(start, end, name, save):
    x = np.arange(0, start.size)
    #popt, pcov = curve_fit(lambda t, a, b: a * np.exp(-b * t), x, start)
    popt, pcov = curve_fit(lambda t, a, b: a * np.exp(-t/b), x, start)
    a = popt[0]
    b = popt[1]
    #c = popt[2]
    c = 0
    x_fitted_curve_fit_start = x#np.linspace(np.min(x), np.max(x), 100)
    #y_fitted_curve_fit_start = a * np.exp(-b * x_fitted_curve_fit_start)+c
    y_fitted_curve_fit_start = a * np.exp(-x_fitted_curve_fit_start/b)+c
    #print(a, b, c)
    print("a = ", a)
    print("b = ", b)
    print("xi = 1/b = ", 1/b)
    #print(pcov)
    perr = np.sqrt(np.diag(pcov))
    print("SE(a): ", perr[0])
    print("SE(b): ", perr[1])
    print(perr)
    #popt, pcov = curve_fit(lambda t, a, b: a * np.exp(-b * t), x, end)
    popt, pcov = curve_fit(lambda t, a, b: a * np.exp(-t/b), x, end)
    a = popt[0]
    b = popt[1]
    print("c_l: ", 1/b)
    #c = popt[2]
    c = 0
    x_fitted_curve_fit_end = x#np.linspace(np.min(x), np.max(x), 100)
    #y_fitted_curve_fit_end = a * np.exp(-b * x_fitted_curve_fit_end)+c
    y_fitted_curve_fit_end = a * np.exp(-x_fitted_curve_fit_end/b)+c
    #print(a, b, c)
    print("a = ", a)
    print("b = ", b)
    print("xi = 1/b = ", 1/b)
    #print(pcov)
    perr = np.sqrt(np.diag(pcov))
    print("SE(a): ", perr[0])
    print("SE(b): ", perr[1])
    print(perr)
    print("mase of start:",mase(start, y_fitted_curve_fit_start))
    print("mase of end:", mase(end, y_fitted_curve_fit_end))
    
    fig, ax = plt.subplots(figsize=(16, 12))
    ax.plot(x, start, label="zippering favoured correlation discrete", marker='o', linewidth=3, markersize=12)  # Plot some data on the axes.
    ax.plot(x, end, label="high crossover correlation discrete", marker='o', linewidth=3, markersize=12)  # Plot some data on the axes.
    ax.plot(x, y_fitted_curve_fit_start, label="zippering favoured correlation fit", marker='', linewidth=3, markersize=1)  # Plot some data on the axes.
    ax.plot(x, y_fitted_curve_fit_end, label="high crossover correlation fit", marker='', linewidth=3, markersize=1)  # Plot some data on the axes.
    ax.legend(loc=2, prop={'size': 18});  # Add a legend.

    ax.set_xlabel('Distance', fontsize=40, fontweight="bold")  # Add an x-label to the axes.
    plt.xticks(fontsize=32, fontweight="bold")
    ax.set_ylabel('Correlation', fontsize=40, fontweight="bold")  # Add a y-label to the axes.
    plt.yticks(fontsize=32, fontweight="bold")
    ax.set_title("Correlation vs Distance", fontsize=40, fontweight="bold", pad=22)  # Add a title to the axes.
    leg = ax.legend(loc='upper right', prop={'size': 24, 'weight': 'bold'});  # Add a legend.
    leg.set_title('Correlation Functions and Fits',prop={'size':32, 'weight': 'bold'})

    #plt.ylim([0, 1])
    plt.xlim([0,40])
    #fig.savefig('test.eps', bbox_inches = "tight")
    if(save):
        fig.savefig(name+'_similarity_fit.eps', bbox_inches = "tight")
    
    #plt.plot(x, fit_start, "b-")
    #plt.plot(x, fit_end, "r-")
    plt.show()
```
